```python
import os
import random
import requests
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt
from tracker import Tracker
import torch
import numpy as np
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
litter_model = YOLO(r"C:\Users\User\Desktop\yolov9_selfbest.onnx")
video_path = "C:\\Users\\User\\Desktop\\dynamic_video.mp4"
output_path = 'C:\\Users\\User\\Desktop\\test_video\\output_dynamic_video2.mp4'
car_model=YOLO('yolov8n.pt').to(device)

cap = cv2.VideoCapture(video_path)

ret, frame = cap.read()
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))


class_id_to_ignore=3
tracker = Tracker()

colors = [(random.randint(0, 255), random.randint(0, 255), random.randint(0, 255)) for j in range(10)]
frame_count=0
detection_threshold = 0.5
id_status={}
while ret:
    frame_count+=1
    results = litter_model(frame)
    print(frame_count)
    for result in results:
        detections = []
        for r in result.boxes.data.tolist():
            x1, y1, x2, y2, score, class_id = r
            x1 = int(x1)
            x2 = int(x2)
            y1 = int(y1)
            y2 = int(y2)
            class_id = int(class_id)
            if class_id != class_id_to_ignore and score > detection_threshold:
                detections.append([x1, y1, x2, y2, score])

        tracker.update(frame, detections,frame_count)
        for track in tracker.tracks:
            
            bbox = track.bbox
            x1, y1, x2, y2 = bbox
            track_id = track.track_id
            id_to_check=track_id
            detect=track.detect
            
            if id_to_check not in id_status:
                id_status[id_to_check]={'status':False,'plate_number':None,'class_id':class_id}
                
           
            
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (colors[track_id % len(colors)]), 3)
            cv2.putText(frame, f"{class_id}-{track_id}", (int(x1) + 10, int(y1) - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 1, cv2.LINE_AA)
            
            if detect and not id_status[track_id]['status']:
                first_frame_central_x=track.first_frame_central_x
                first_frame_central_y=track.first_frame_central_y
                first_frame=track.first_frame
                vehicles_results = car_model(first_frame, classes=[2,3,5])
                vehicle_data=[]
                for vehicles_result in vehicles_results:
                    for vehicle_r in vehicles_result.boxes.data.tolist():
                        vehicle_x1,vehicle_y1,vehicle_x2,vehicle_y2,vehicle_score,vehicle_class=vehicle_r
                        center_x=(vehicle_x1+vehicle_x2)/2
                        center_y=(vehicle_y1+vehicle_y2)/2
                        vehicle_info = {
                        'bbox': (vehicle_x1, vehicle_y1, vehicle_x2, vehicle_y2),
                        'center': (center_x, center_y),
                        'score': vehicle_score,
                        'class': vehicle_class
                        }
                        vehicle_data.append(vehicle_info)
               
                min_distance = float('inf')
                closest_vehicle_info = None
                for vehicle in vehicle_data:
                    distance = np.sqrt((vehicle['center'][0] - first_frame_central_x) ** 2 + (vehicle['center'][1] - first_frame_central_y) ** 2)
                if distance < min_distance:
                    min_distance = distance
                    closest_vehicle_info = vehicle
                if closest_vehicle_info:
                    closest_vehicle_y1 = int(closest_vehicle_info['bbox'][1])
                    closest_vehicle_y2 = int(closest_vehicle_info['bbox'][3])
                    closest_vehicle_x1 = int(closest_vehicle_info['bbox'][0])
                    closest_vehicle_x2 = int(closest_vehicle_info['bbox'][2])

                    vehicle_image = first_frame[closest_vehicle_y1:closest_vehicle_y2, closest_vehicle_x1:closest_vehicle_x2]
                    plt.imshow(vehicle_image)  
                    plt.axis('off')  # 关闭坐标轴显示
                    plt.show()

                    success, image_jpg = cv2.imencode('.jpg', vehicle_image)
                    if not success:
                        raise Exception("Image encoding failed")
                    files = {'upload': image_jpg.tobytes()}
                    response = requests.post(
                    'https://api.platerecognizer.com/v1/plate-reader/',
                    files=files,
                    headers={'Authorization': 'Token e2fe8cc56b9fb5924d15150ffebb260ba798fc2d'}
                    )

                    if response.status_code == 201:
                        data = response.json()  
                        plates = [result['plate'] for result in data.get('results', [])]  
                        print("Detected license plates:", plates)
                        id_status[track_id]['status']=True
                        id_status[track_id]['plate_number']=plates
                    else:
                        print("Failed to get valid response, status code:", response.status_code)
                
                
                cv2.putText(frame, f"ID_name:{track_id} litter dectect!", (int(x1) +10, int(y1) - 30), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 1, cv2.LINE_AA)
    out.write(frame)
    # Display the annotated frame
    cv2.imshow("YOLOv8 Tracking", frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord("q"):
         break 
    ret, frame = cap.read()
for litter_trackid ,info in id_status.items():
    if info['status']:
        if info['class_id']==0:
            print(f"抓到亂丟垃圾車主，車牌號碼:{info['plate_number']}，垃圾編號為:{litter_trackid}")
        elif info['class_id']==1:
            print(f"抓到亂丟垃圾車主，車牌號碼:{info['plate_number']}，垃圾編號為:{litter_trackid}")
        elif info['class_id']==2:
            print(f"抓到亂丟垃圾車主，車牌號碼:{info['plate_number']}，垃圾編號為:{litter_trackid}")
        elif info['class_id']==3:
            print(f"抓到亂丟垃圾車主，車牌號碼:{info['plate_number']}，垃圾編號為:{litter_trackid}")
        elif info['class_id']==5:
            print(f"抓到亂丟垃圾車主，車牌號碼:{info['plate_number']}，垃圾編號為:{litter_trackid}")
            

cap.release()
out.release() 
cv2.destroyAllWindows()
```
